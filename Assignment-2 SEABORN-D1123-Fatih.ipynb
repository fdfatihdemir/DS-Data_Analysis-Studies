## 2: Plot 12 most dangerous cities

```python
city_freq = kill.city.value_counts()[:12]
city_freq
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(x=city_freq.index,y=city_freq.values,data=kill)
ax= plt.xticks(rotation=45)
ax=plt.title('Most dangerous cities',color = 'blue',fontsize=15)
```

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x=city_freq.index,data=kill)
ax= plt.xticks(rotation=45)
ax=plt.title('Most dangerous cities',color = 'blue',fontsize=15)
```

## 3: Plot 20 most dangerous states

```python
# with seaborn countplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x=state_freq.index, data=kill)
```
